## Q1

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
```

```python
# Create color maps(one color for each class)
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
# import some data to play with
iris = datasets.load_iris()
X_all = iris.data
y = iris.target
print(X_all.shape)
print(y.shape)
```

```python
# TODO: using scatter plots, find which combination of 2 features will work better
SEPAL_LENGTH = 0
SEPAL_WIDTH = 1
PETAL_LENGTH = 2
PETAL_WIDTH = 3
X = X_all[:, [PETAL_WIDTH, PETAL_LENGTH]] # Use 2 features only
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df
```

```python
# we create an instance of Neighbours Classifier and fit the data.
n_neighbors = 15
clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
clf.fit(X, y)
y_pred = clf.predict(X)
print(y_pred.shape)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
plt.figure(figsize=(10,10))
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# visualize results
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,  edgecolor='k', s=50)
plt.scatter(X[:, 0], X[:, 1], c=y_pred,      cmap=cmap_light, edgecolor='k', s=100, marker='x')
plt.show()
```

Les rÃ©sultats entre "y" et "y_pred" sont quasiment identiques car notre prÃ©diction est basÃ©e sur les valeurs rÃ©elles provenant de "y". De ce fait, en changeant la constante "n_neighbors", les rÃ©sultats restent les mÃªmes. On parle alors de "Overfitting".

## Q3


```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original', data_home='./')

print(mnist.data.shape)
print(mnist.target.shape)
print(np.unique(mnist.target))
```

```python
from sklearn.model_selection import train_test_split

X = mnist.data[::100]
y = mnist.target[::100]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

n_neighbors = 5
clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(y_pred.shape)
print(y_test.shape)
print(y_pred)
print(y_test)
```

On peut voir que la longueur est la mÃªme entre y_pred et y_test et que les donnÃ©es sont quasiment identiques.

## Q4

```python
import sklearn.metrics as metrics
score1 = metrics.accuracy_score(y_test, y_pred)
score2 = metrics.precision_score(y_test, y_pred, average=None)
score3 = metrics.classification_report(y_test, y_pred)
print('score1 =', score1)
print('score2 =', score2)
print('score3 =\n', score3)
```

On peut remarquer que la prÃ©cision de la prÃ©diction est de 87%.
